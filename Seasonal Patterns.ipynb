```python
# By seasonality, we mean the presence of patterns that occur at regular intervals (shorter than a year)
# For this example we'll be using unemplyement statistics from Nasdaqdatalink
```

```python
import pandas as pd
import nasdaqdatalink
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Download and display unemployment data from Nasdaq Data Link:
df = (
 nasdaqdatalink.get(dataset="FRED/UNRATENSA",
 start_date="2014-01-01",
 end_date="2022-12-31")
 .rename(columns={"Value": "unemp_rate"})
)
df.plot(title="Unemployment rate in years 2014-2022")
```

```python
# we can already spot some seasonal (repeating) patterns, for example, each year unemployment seems to be highest in January
```

```python
# Create new columns with year and month:
df["year"] = df.index.year
df["month"] = df.index.strftime("%b") # tells us the month based on the index
```

```python
# Create the seasonal plot:
sns.lineplot(data=df,
 x="month",
 y="unemp_rate",
 hue="year",
 style="year",
 legend="full",
 palette="colorblind")
plt.title("Unemployment rate - Seasonal plot")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
```
