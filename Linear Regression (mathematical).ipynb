```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('test.csv')
```

```python
plt.scatter(data['Study Time'],data['Scores'])
plt.show()
```

```python
# Basic Loss Function
def loss_function(m,b,points):
  total_error = 0
  for i in range(len(points)):
    x = points.iloc[i]['Study Time']
    y = points.iloc[i]['Scores']
    total_error += (y - (m * x + b)) ** 2
  total_error / float(len(points))

# This function tells us how off we are from the actual result
```

```python
# Gradiant Descent
def gradient_descent(m_now, b_now, points, L): # L is the learning rate
  m_gradient = 0
  b_gradient = 0

  n = len(points)

  for i in range(n):
    x = points.iloc[i]['Study Time']
    y = points.iloc[i]['Scores']

    m_gradient += -(2/n) * x * (y - (m_now * x * b_now))
    b_gradient += -(2/n) * (y - (m_now * x * b_now))

  m = m_now - m_gradient * L # opposite direction with a learning rate of L
  b = b_now - b_gradient * L
  return m,b
```

```python
# Executing the linear regression

m = 0
b = 0
L = 0.0001
epochs = 100 # iterations
```

```python
for i in range(epochs):
  m, b = gradient_descent(m,b,data,L) # we will keep getting better and better at getting the perfect estimate of m and b

print(m,b)
```
